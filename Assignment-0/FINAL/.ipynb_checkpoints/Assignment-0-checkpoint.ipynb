## Converting video to images

```python
def video_description(filename):
    cap = cv2.VideoCapture(filename)
    fps = cap.get(cv2.CAP_PROP_FPS)
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    duration = frame_count/fps
    print(f"\nVideo source: {filename}")
    print(f"FPS: {fps}")
    print(f"Total frames: {frame_count}")
    print(f"Total Duration: {duration} seconds")
    print(f"-----------------------------")
    
video_description('Data/Q1.mp4')
```

```python
def convert_video_to_frames(input_path,output_path):
    vid = cv2.VideoCapture(input_path)
    count = 0
    while True:
        success, img = vid.read()
        if success == True:
            count = count+1
            cv2.imwrite(output_path+"/frame"+str(count)+".jpg", img)
        else:
            break
    
!rm -rf Results/Q1/Frames/*
print("Directory content before conversion:")
!ls -la Results/Q1/Frames

convert_video_to_frames('Data/Q1.mp4','Results/Q1/Frames')
    
print("\nDirectory content after conversion:")
!ls -la Results/Q1/Frames
```

## Converting images to video
lorem ipsum lawda lassan

24fps â€“This is the standard for movies and TV shows, and it was determined to be the minimum speed needed to capture video while still maintaining realistic motion. Even if a film is shot at a higher frame rate, it's often produced and displayed at 24fps. Most feature films and TV shows are shot and viewed at 24 fps.

```python
def convert_frames_to_video(input_path,output_path,fps):
    all_frames = []
    files = os.listdir(input_path)
    files.sort(key = lambda x: int(x[5:-4]))
    for i in range(len(files)):
        filename = input_path + files[i]
        img = cv2.imread(filename)
        height, width, layers = img.shape
        size = (width,height)
        all_frames.append(img)
    out = cv2.VideoWriter(output_path,cv2.VideoWriter_fourcc(*'DIVX'), fps, size)
    for i in range(len(all_frames)):
        out.write(all_frames[i])
    out.release()

frame_rates = [12, 25, 30, 45]
input_path= './Results/Q1/Frames/'

!rm -rf Results/Q1/Videos/*
print("Directory content before conversion:")
!ls -la Results/Q1/Videos

for fps in frame_rates:    
    output_path = './Results/Q1/Videos/Combined-'+str(fps)+'.avi'
    convert_frames_to_video(input_path, output_path, fps)
    video_description(output_path)    
    
print("\nDirectory content after conversion:")
!ls -la Results/Q1/Videos
```

## Chroma Keying:
lorem ipsum lawda lassan

```python
def chroma_key(filename1, filename2, output_path):

    video1 = cv2.VideoCapture(filename1)
    video2 = cv2.VideoCapture(filename2)

    cnt = 0

    while True:

        cnt += 1

        ret1, frame1 = video1.read()
        ret2, frame2 = video2.read()

        ret = ret1 and ret2

        if ret == True:

            frame1 = cv2.resize(frame1, (640, 360))
            frame2 = cv2.resize(frame2, (640, 360))
            frame1_copy = np.copy(frame1)
            l_bound = np.array([160, 160, 160])
            u_bound = np.array([255, 255, 255])
            mask = cv2.inRange(frame1_copy, l_bound, u_bound)
            masked_image = np.copy(frame1_copy)
            masked_image[mask != 0] = [0, 0, 0]
            background = frame2[0:360, 0:640]
            background[mask == 0] = [0, 0, 0]
            final_image = background + masked_image
            cv2.imshow("output", final_image)
            cv2.imwrite(output_path+"/frame"+str(cnt)+".jpg", final_image)

            if cv2.waitKey(10) == 27:
                break

        else:
            break

    video1.release()
    video2.release()
    cv2.destroyAllWindows()
```
