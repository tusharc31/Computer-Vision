## Converting video to images

We know that a video is just images being played sequentially with high speed. To convert a video to images, all I have done is use `read()` method of opencv and read all the frames one by one and stored them in a directory. All the images have been shown to be stored and some random frames have with visualized as well.

```python
# A utility function to get basic info about 
# the video such as FPS, length, etc
def video_description(filename):
    cap = cv2.VideoCapture(filename)
    fps = cap.get(cv2.CAP_PROP_FPS)
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    duration = frame_count/fps
    print(f"\nVideo source: {filename}")
    print(f"FPS: {fps}")
    print(f"Total frames: {frame_count}")
    print(f"Total Duration: {duration} seconds")
    print(f"-----------------------------")
    
video_description('Data/Q1.mp4')
```

```python
def convert_video_to_frames(input_path,output_path):
    vid = cv2.VideoCapture(input_path)
    count = 0
    while True:
        success, img = vid.read()
        if success == True:
            count = count+1
            cv2.imwrite(output_path+"/frame"+str(count)+".jpg", img)
            if count==1 or count==41 or count==81:
                print(f"\nSample frame: number {count}")
                plt.imshow(img)
                plt.show()
        else:
            break
    
!rm -rf Results/Q1/Frames/*
print("Directory content before conversion:")
!ls -la Results/Q1/Frames

convert_video_to_frames('Data/Q1.mp4','Results/Q1/Frames')
    
print("\nDirectory content after conversion:")
!ls -la Results/Q1/Frames
```

## Converting images to video

In this question, I have used the same set of images generated in the previous part. To convert these images to a video I have used the `VideoWriter` method from opencv. The most important part of this question is the frame rate. We know FPS = Total frames / duration. Hence if we change the FPS, the duration of the generated video will change as the total number of frames are constant. Based on the frame rate, these are the following cases I have tried.

* FPS = 30. This frame rate is same as of the original video and hence I was able to generate a video with same speed and duration as the original one.
* FPS = 45. Having a frame rate higher than the frame rate of the original video made the new video faster and hence of lesser duration then the original video.
* FPS = 25. Having a frame rate lower than the frame rate of the original video made the new video slower and hence of more duration then the original video.
* FPS = 12. We know human eyes process a sequence of images as a realistic motion video if the frame rate is higher than 24. So I have also generated a video with frame rate of 12 and here I can feel the motion between the frames is not smooth enough/realistic.

All the videos have been shown to be written in a directory and basic information for them has been printed.

```python
def convert_frames_to_video(input_path,output_path,fps):
    all_frames = []
    files = os.listdir(input_path)
    files.sort(key = lambda x: int(x[5:-4]))
    for i in range(len(files)):
        filename = input_path + files[i]
        img = cv2.imread(filename)
        height, width, layers = img.shape
        size = (width,height)
        all_frames.append(img)
    out = cv2.VideoWriter(output_path,cv2.VideoWriter_fourcc(*'DIVX'), fps, size)
    for i in range(len(all_frames)):
        out.write(all_frames[i])
    out.release()

frame_rates = [12, 25, 30, 45]
input_path= './Results/Q1/Frames/'

!rm -rf Results/Q1/Videos/*
print("Directory content before conversion:")
!ls -la Results/Q1/Videos

for fps in frame_rates:    
    output_path = './Results/Q1/Videos/Combined-'+str(fps)+'.avi'
    convert_frames_to_video(input_path, output_path, fps)
    video_description(output_path)    
    
print("\nDirectory content after conversion:")
!ls -la Results/Q1/Videos
```

## Chroma Keying:

For this question, we had to use two videos, and via chroma keying we had to add foreground of one video on a background from another video. I attempted this problem twice and my challenges and learnings have been explained below:

* In the first attempt, I used a `5 second duration at 25 FPS` video, and another video of me shot on my phone which was roughly `5 seconds long at 120 FPS`. Hence both videos had unequal frames (5x25 and 5x120 respectively). I read both the videos frame by frame, and till I could read frames, I masked pixels of image from my video with RGB values in the range [160, 160, 160] to [255, 255, 255]. This range was found by logical trial and error (the background in my video was white). After this, I added it to the background image from the second video. However, it was very noticiable that both the videos were not in sync at all. My video was going very slow compared to the background, and this was beacuse two videos of unequal frame rates were being played at the same FPS.

* In the second attempt, I preproccesed my video, so now it was roughly `5 seconds long at 24 FPS`. I performed the preprocessing by converting the video to frames, and then converting it back to a video by choosing 1 in every 5 frames. I did this since I wanted to bring the FPS of my video closer to the background video. Hence, I changed the FPS from 120 to 24 and reduced number of frames by 5 times. Rest of the procedure was carried in the same way as before and this time both the videos were in sync.


```python
def chroma_key(filename1, filename2, output_path):

    video1 = cv2.VideoCapture(filename1)
    video2 = cv2.VideoCapture(filename2)

    cnt = 0

    while True:

        cnt += 1

        ret1, frame1 = video1.read()
        ret2, frame2 = video2.read()

        ret = ret1 and ret2

        if ret == True:

            frame1 = cv2.resize(frame1, (640, 360))
            frame2 = cv2.resize(frame2, (640, 360))
            frame1_copy = np.copy(frame1)
            l_bound = np.array([160, 160, 160])
            u_bound = np.array([255, 255, 255])
            mask = cv2.inRange(frame1_copy, l_bound, u_bound)
            masked_image = np.copy(frame1_copy)
            masked_image[mask != 0] = [0, 0, 0]
            background = frame2[0:360, 0:640]
            background[mask == 0] = [0, 0, 0]
            final_image = background + masked_image
            cv2.imshow("output", final_image)
            cv2.imwrite(output_path+"/frame"+str(cnt)+".jpg", final_image)
            
            if cnt==10 or cnt==30 or cnt==50:
                frame1 = cv2.cvtColor(frame1, cv2.COLOR_BGR2RGB)
                frame2 = cv2.cvtColor(frame2, cv2.COLOR_BGR2RGB)
                final_image = cv2.cvtColor(final_image, cv2.COLOR_BGR2RGB)
                print("\nSAMPLE FRAMES:")
                plt.imshow(frame1)
                plt.show()
                plt.imshow(frame2)
                plt.show()
                plt.imshow(final_image)
                plt.show()

            if cv2.waitKey(10) == 27:
                break

        else:
            break

    video1.release()
    video2.release()
    cv2.destroyAllWindows()
```

```python
video_description('Data/Q3/vid1-1.mp4')
video_description('Data/Q3/vid2.mp4')

!rm -rf Results/Q3/vid1_1/*
print("\nDirectory content before chroma keying:")
!ls -la Results/Q3/vid1_1

chroma_key('Data/Q3/vid1-1.mp4', 'Data/Q3/vid2.mp4', "Results/Q3/vid1_1")

print("\nDirectory content after chroma keying:")
!ls -la Results/Q3/vid1_1
```
