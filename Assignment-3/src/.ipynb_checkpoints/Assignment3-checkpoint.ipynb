For this assignment you will implement the GrabCut method mentioned in this [paper](https://cvg.ethz.ch/teaching/cvl/2012/grabcut-siggraph04.pdf). It is essentially an iterative version of GraphCut as shown in the figure below. 


The code below takes an input image and follows these steps:
- It requires a bounding box to be drawn by the user to roughly segment out the foreground pixels
- It runs an initial min-cut optimization using the provided annotation
- The result of this optimization gives an initial segmentation 
- To further refine this segmentation, the user provides two kinds of strokes to aid the optimization
    - strokes on the background pixels
    - strokes on the foreground pixels
- The algorithm now utilizes this to refine the original segmentation

You are allowed to use standard GMM libraries for the implementation. For usage of other libraries, please contact the TAs.

You can view this [video](https://www.youtube.com/watch?v=aOqOwM-Qbtg) to get a better idea of the steps involved.

Image segmentation is one exciting application of MRFs. You can further read about other applications of MRFs for Computer Vision [here](https://cedar.buffalo.edu/~srihari/CSE574/Chap8/Ch8-PGM-Undirected/9.5-MRFinCV.pdf).

```python
import numpy as np
import cv2
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt
import igraph as ig
import time
import os
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
NUM_GMM_COMP = 5
GAMMA = 50
NUM_ITERS = 3
UNARY_TYPE = 1
```

```python
class EventHandler:
    """
    Class for handling user input during segmentation iterations 
    """
    
    def __init__(self, flags, img, _types, _alphas, colors):
        
        self.FLAGS = flags
        self.ix = -1
        self.iy = -1
        self.img = img
        self.img2 = self.img.copy()
        self._types = _types
        self._alphas = _alphas
        self.COLORS = colors

    @property
    def image(self):
        return self.img
    
    @image.setter
    def image(self, img):
        self.img = img
        
    @property
    def types(self):
        return self._types

    @types.setter
    def types(self, _types):
        self._types = _types
    
    @property
    def alphas(self):
        return self._alphas

    @alphas.setter
    def alphas(self, _alphas):
        self._alphas = _alphas
    
    @property
    def flags(self):
        return self.FLAGS 
    
    @flags.setter
    def flags(self, flags):
        self.FLAGS = flags
    
    def handler(self, event, x, y, flags, param):

        # Draw the rectangle first
        if event == cv2.EVENT_LBUTTONDOWN:
            if self.FLAGS['rect_over'] == False:
                self.FLAGS['DRAW_RECT'] = True
                self.ix, self.iy = x,y

        elif event == cv2.EVENT_MOUSEMOVE:
            if self.FLAGS['DRAW_RECT'] == True:
                self.img = self.img2.copy()
                cv2.rectangle(self.img, (self.ix, self.iy), (x, y), self.COLORS['BLUE'], 2)
                self.FLAGS['RECT'] = (min(self.ix, x), min(self.iy, y), abs(self.ix - x), abs(self.iy - y))
                self.FLAGS['rect_or_mask'] = 0

        elif event == cv2.EVENT_LBUTTONUP:
            if self.FLAGS['rect_over'] == False:
                self.FLAGS['DRAW_RECT'] = False
                self.FLAGS['rect_over'] = True
                cv2.rectangle(self.img, (self.ix, self.iy), (x, y), self.COLORS['BLUE'], 2)
                self.FLAGS['RECT'] = (min(self.ix, x), min(self.iy, y), abs(self.ix - x), abs(self.iy - y))
                self.FLAGS['rect_or_mask'] = 0

                # initialise types and alphas
                temp = np.zeros(self._types.shape, dtype=np.uint8)
                rect = self.FLAGS['RECT']
                temp[rect[1]:rect[1]+rect[3], rect[0]:rect[0]+rect[2]] = 1
                self._types[temp == 0] = 1
                self._alphas[rect[1]:rect[1]+rect[3], rect[0]:rect[0]+rect[2]] = 1

        
        # Draw strokes for refinement 

        if event == cv2.EVENT_LBUTTONDOWN:
            if self.FLAGS['rect_over'] == True:
                self.FLAGS['DRAW_STROKE'] = True
                cv2.circle(self.img, (x,y), 3, self.FLAGS['value']['color'], -1)
                cv2.circle(self._alphas, (x,y), 3, self.FLAGS['value']['val'], -1)
                cv2.circle(self._types, (x,y), 3, 1, -1)

        elif event == cv2.EVENT_MOUSEMOVE:
            if self.FLAGS['DRAW_STROKE'] == True:
                cv2.circle(self.img, (x, y), 3, self.FLAGS['value']['color'], -1)
                cv2.circle(self._alphas, (x,y), 3, self.FLAGS['value']['val'], -1)
                cv2.circle(self._types, (x,y), 3, 1, -1)

        elif event == cv2.EVENT_LBUTTONUP:
            if self.FLAGS['DRAW_STROKE'] == True:
                self.FLAGS['DRAW_STROKE'] = False
                cv2.circle(self.img, (x, y), 3, self.FLAGS['value']['color'], -1)
                cv2.circle(self._alphas, (x,y), 3, self.FLAGS['value']['val'], -1)
                cv2.circle(self._types, (x,y), 3, 1, -1)
```

```python
def run(filename, gt_name, n_components=NUM_GMM_COMP, gamma=GAMMA, lamda=9*GAMMA,
        num_iters=NUM_ITERS, tol=1e-3, connect_diag=True, unary_type=UNARY_TYPE, plot_flg=True):
    
    gt_alphas = plt.imread(gt_name)
    
    for i in range(gt_alphas.shape[0]):
        for j in range(gt_alphas.shape[1]):
            if gt_alphas[i][j]==255:
                gt_alphas[i][j] = 1
    
    """
    Main loop that implements GrabCut. 
    
    Input
    -----
    filename (str) : Path to image
    """
    
    COLORS = {
    'BLACK' : [0,0,0],
    'RED'   : [0, 0, 255],
    'GREEN' : [0, 255, 0],
    'BLUE'  : [255, 0, 0],
    'WHITE' : [255,255,255]
    }

    DRAW_BG = {'color' : COLORS['BLACK'], 'val' : 0}
    DRAW_FG = {'color' : COLORS['WHITE'], 'val' : 1}

    FLAGS = {
        'RECT' : (0, 0, 1, 1),
        'DRAW_STROKE': False,         # flag for drawing strokes
        'DRAW_RECT' : False,          # flag for drawing rectangle
        'rect_over' : False,          # flag to check if rectangle is  drawn
        'rect_or_mask' : -1,          # flag for selecting rectangle or stroke mode
        'value' : DRAW_FG,            # drawing strokes initialized to mark foreground
    }

    img = cv2.imread(filename)
    img2 = img.copy()
    types = np.zeros(img.shape[:2], dtype = np.uint8)  # whether a pixel is known or unknown
    alphas = np.zeros(img.shape[:2], dtype = np.uint8) # mask is a binary array with : 0 - background pixels
                                                       #                               1 - foreground pixels 
    output = np.zeros(img.shape, np.uint8)             # output image to be shown

    # Input and segmentation windows
    cv2.namedWindow('Input Image')
    
    EventObj = EventHandler(FLAGS, img, types, alphas, COLORS)
    cv2.setMouseCallback('Input Image', EventObj.handler)
    cv2.moveWindow('Input Image', img.shape[1] + 10, 90)

    while(1):
        
        img = EventObj.image
        types = EventObj.types
        alphas = EventObj.alphas
        FLAGS = EventObj.flags
        cv2.imshow('Segmented image', output)
        cv2.imshow('Input Image', img)
        
        k = cv2.waitKey(1)

        # key bindings
        if k == 27:
            # esc to exit
            break
        
        elif k == ord('0'): 
            # Strokes for background
            FLAGS['value'] = DRAW_BG
        
        elif k == ord('1'):
            # FG drawing
            FLAGS['value'] = DRAW_FG
        
        elif k == ord('r'):
            # reset everything
            FLAGS['RECT'] = (0, 0, 1, 1)
            FLAGS['DRAW_STROKE'] = False
            FLAGS['DRAW_RECT'] = False
            FLAGS['rect_or_mask'] = -1
            FLAGS['rect_over'] = False
            FLAGS['value'] = DRAW_FG
            img = img2.copy()
            types = np.zeros(img.shape[:2], dtype = np.uint8) 
            alphas = np.zeros(img.shape[:2], dtype = np.uint8)
            EventObj.image = img
            EventObj.types = types
            EventObj.alphas = alphas
            output = np.zeros(img.shape, np.uint8)
        
        elif k == 13: 
            # Press carriage return to initiate segmentation
            
            alphas = grab_cut(img2, types, alphas, n_components, gamma, lamda, num_iters, tol, connect_diag, unary_type, gt_alphas)
            EventObj.alphas = alphas
        
        EventObj.flags = FLAGS
        mask2 = np.where((alphas == 1), 255, 0).astype('uint8')
        output = cv2.bitwise_and(img2, img2, mask = mask2)
        
    if plot_flg==True:
        plt.imshow(output)
```

```python
def prob_distribution(d, gmm_model):
    
    gmm_alphas = gmm_model.alphasict
    gmm_cov = gmm_model.covariances_
    gmm_means = gmm_model.means_
    gmm_weights = gmm_model.weights_
    idx = gmm_alphas(d.reshape(1, -1))
    
    val1 = np.log(np.linalg.det(gmm_cov[idx[0]]))
    val2 = np.dot(np.dot(np.transpose(d - gmm_means[idx[0]]), np.linalg.inv(gmm_cov[idx[0]])), d - gmm_means[idx[0]])
    return (val1/2 + val2/2 - np.log(gmm_weights[idx[0]]))

def segmentation_graph(img, types, alphas, fg_gmm, bg_gmm, beta, gamma, lamda, connect_diag, unary_type):
    
    d1 = img.shape[0]
    d2 = img.shape[1]
    edges = []
    weights = []
    fg_D = np.zeros(img.shape[:-1])
    bg_D = np.zeros(img.shape[:-1])
    
    if unary_type == 1:
        fg_D = -1*fg_gmm.score_samples(img.reshape((-1, img.shape[-1]))).reshape(img.shape[:-1])
        bg_D = -1*bg_gmm.score_samples(img.reshape((-1, img.shape[-1]))).reshape(img.shape[:-1])

    else:
        for i in range(d1):
            for j in range(d2):
                fg_D[i][j] = prob_distribution(img[i][j], fg_gmm)
                bg_D[i][j] = prob_distribution(img[i][j], bg_gmm)

    S = d1*d2
    T = d1*d2+1                
    graph = ig.Graph(directed=False)
    graph.add_vertices(T+1)    
    
    for i in range(d1):
        for j in range(d2):
            # pair-wise potentials
            if i+1 < d1:
                edges.append(((d2*i)+j, (d2*(i+1))+j))
                weights.append(gamma*np.exp(- beta*(np.linalg.norm(img[i, j] - img[i+1, j])**2)))
            if j+1 < d2:
                edges.append(((d2*i)+j, (d2*i)+j+1))
                weights.append(gamma*np.exp(- beta*(np.linalg.norm(img[i, j] - img[i, j+1])**2)))
            if connect_diag:
                if i+1 < d1 and j+1 < d2:
                    edges.append(((d2*i)+j, (d2*(i+1))+j+1))
                    weights.append(gamma*np.exp(- beta*(np.linalg.norm(img[i, j] - img[i+1, j+1])**2)))
                if i+1 < d1 and j > 0:
                    edges.append(((d2*i)+j, (d2*(i+1))+j-1))
                    weights.append(gamma*np.exp(- beta*(np.linalg.norm(img[i, j] - img[i+1, j-1])**2)))
            # unary potentials
            if types[i, j] == 0: # If node is not fixed yet
                edges.append(((d2*i)+j, S))
                edges.append(((d2*i)+j, T))
                weights.append(bg_D[i, j])
                weights.append(fg_D[i, j])
            else: # node is fixed
                if alphas[i, j] == 1:
                    edges.append(((d2*i)+j, S))
                    weights.append(lamda)
                else:
                    edges.append(((d2*i)+j, T))
                    weights.append(lamda)
    
    graph.add_edges(edges, attributes={'weight': weights})
    new_alphas = np.zeros(img.shape[:2], dtype=np.uint8)
    cut = graph.st_mincut(S, T, capacity='weight')
    stcost = cut.value
    
    if S in cut.partition[1]:
        for v in cut.partition[1]:
            if v not in (S, T):
                new_alphas[((int(v / d2)), (v%d2))] = 1
    
    else:
        for v in cut.partition[0]:
            if v not in (S, T):
                new_alphas[((int(v / d2)), (v%d2))] = 1
    
    return [stcost, new_alphas]


def calc_beta(img, connect_diag):
    
    d1 = img.shape[0]
    d2 = img.shape[1]
    beta = 0
    
    if connect_diag:
        for i in range(d1):
            for j in range(d2):
                if i+1 < d1: beta = beta+(np.linalg.norm(img[i, j] - img[i+1, j])**2)
                if j+1 < d2: beta = beta+(np.linalg.norm(img[i, j] - img[i, j+1])**2)
                if i+1 > d1 and j+1 < d2: beta = beta+(np.linalg.norm(img[i, j] - img[i+1, j+1])**2)
                if i+1 < d1 and j > 0: beta = beta+(np.linalg.norm(img[i, j] - img[i+1, j-1])**2)
        beta = beta / (2*d1*d2 - d1 - d2)

    else:
        for i in range(d1):
            for j in range(d2):
                if i+1 < d1: beta = beta+(np.linalg.norm(img[i, j] - img[i+1, j])**2)
                if j+1 < d2: beta = beta+(np.linalg.norm(img[i, j] - img[i, j+1])**2)
        beta = beta / (4*d1*d2 - 3*d1 - 3*d2+2)
    
    return 1/(2*beta)

def get_accuracy(alphas, gt):
    correct_pix = (alphas == gt).astype(np.float32).sum()
    total_pix = alphas.shape[0]*alphas.shape[1]
    return correct_pix/total_pix

def get_jaccard(alphas, gt):
    # for foreground
    fg_num = np.sum(np.bitwise_and(alphas, gt))
    fg_den = np.sum(np.bitwise_or(alphas, gt))
    # for background
    bg_num = np.sum(np.bitwise_and(1-alphas, 1-gt))
    bg_den = np.sum(np.bitwise_or(1-alphas,1-gt))
    return (fg_num/bg_num + bg_num/bg_den)/2

def get_dice_sim(alphas, gt):
    # for foreground
    fg_num = 2*np.sum(np.bitwise_and(alphas, gt))
    fg_den = 2*alphas.shape[0]*alphas.shape[1]
    # for foreground
    bg_num = 2*np.sum(np.bitwise_and(1-alphas, 1-gt))
    bg_den = 2*alphas.shape[0]*alphas.shape[1]
    return (fg_num/bg_num + bg_num/bg_den)/2

def grab_cut(img_, types_, alphas_, n_components, gamma, lamda, num_iters, tol, connect_diag, unary_type, gt_alphas):
    
    print("\nStarted grabcut iteration ...")
    starting_time = time.process_time()
    img = img_.copy().astype(np.float32)
    prev_cut_cost = None
    types = types_.copy()
    alphas = alphas_.copy()
    beta = calc_beta(img, connect_diag)
    
    for itr in range(num_iters):
        fg_gmm = GaussianMixture(n_components=n_components)
        fg_gmm.fit(img[alphas == 1].reshape((-1, img.shape[-1])))
        bg_gmm = GaussianMixture(n_components=n_components)
        bg_gmm.fit(img[alphas == 0].reshape((-1, img.shape[-1])))
        [cut_cost, alphas] = segmentation_graph(img, types, alphas, fg_gmm, bg_gmm, beta, gamma, lamda, connect_diag, unary_type)
        if prev_cut_cost != None and abs(prev_cut_cost - cut_cost) < tol:
            print("Reached tolerance!")
            break
        prev_cut_cost = cut_cost
    
    print(f"Finished iteration in {int(time.process_time() - starting_time)} seconds")
    print(f"Cost of cut was {cut_cost} units")
    print(f"Accuracy: {get_accuracy(alphas, gt_alphas)}")
    print(f"Jaccard similarity: {get_jaccard(alphas, gt_alphas)}")
    print(f"Dice similarity coefficient: {get_dice_sim(alphas, gt_alphas)}")
    print("- - - - - - - - - - - - - - - - -")
    return alphas
```

## Analysis

**1) The number of iterations of GMM updating and energy minimization**

* Doing only 1 iteration

```python
try: run('../images/llama.jpg', '../ground_truth/llama.bmp', num_iters=1)
finally: cv2.destroyAllWindows()
```

* Doing 5 iterations

```python
try: run('../images/llama.jpg', '../ground_truth/llama.bmp', num_iters=5)
finally: cv2.destroyAllWindows()
```

It is observed that as we increase the number of iterations we get a more refined segmentation of the foreground object until it converges. So the metrics slightly improve.

**2) Different ways to represent probabilities other than GMMs. 4-neighborhood or 8-neighborhood in your pairwise term.**

* Testing with 4-neighborhood

```python
try: run('../images/elefant.jpg', '../ground_truth/elefant.bmp', connect_diag=False)
finally: cv2.destroyAllWindows()
```

* Testing with 8-neighborhood

```python
try: run('../images/elefant.jpg', '../ground_truth/elefant.bmp', connect_diag=True)
finally: cv2.destroyAllWindows()
```

Using a 8 neighbourhood gives a more continuous segmentation as the diagonal pixels also influence the segmentation. The four-connectivity has rougher edges. This happens because whether or not two diagonally adjacent pixels are categorised similarly has no affect on the energy giving the edges a staircase-like look.

**3) The number of mixture components in your GMM**

* Testing with 10 GMMs

```python
try: run('../images/person8.jpg', '../ground_truth/person8.bmp', n_components=10)
finally: cv2.destroyAllWindows()
```

* Testing with 5 GMMs

```python
try: run('../images/person8.jpg', '../ground_truth/person8.bmp', n_components=5)
finally: cv2.destroyAllWindows()
```

* Testing with 5 GMMs and strokes

```python
try: run('../images/person8.jpg', '../ground_truth/person8.bmp', n_components=5)
finally: cv2.destroyAllWindows()
```

Higher number of Gaussian Mixture components can learning and representing a wider range of colors for the foreground and background. Too less components is very low can won't be complex enough to represent the foreground and background distribution. But if the components are too high then there can be additional background pixels present in the segmentation of foreground object. Adding manual strokes help improve the models for foreground and background and hence the GMMs improve.

**4) The choice of gamma**

* Testing with gamma=1

```python
try: run('../images/llama.jpg', '../ground_truth/llama.bmp', gamma=1)
finally: cv2.destroyAllWindows()
```

* Testing with gamma = 50

```python
try: run('../images/llama.jpg', '../ground_truth/llama.bmp', gamma=50)
finally: cv2.destroyAllWindows()
```

The segmentation becomes better and accuracy increases on increasing the value of gamma. When the gamma value was too small, the pairwise costs are not high enough and we can see that the segmentation regions are not continuous. Since the pairwise cost increases on increasing gamma the segmentated regions become continuous.
