For this assignment you will implement the GrabCut method mentioned in this [paper](https://cvg.ethz.ch/teaching/cvl/2012/grabcut-siggraph04.pdf). It is essentially an iterative version of GraphCut as shown in the figure below. 


The code below takes an input image and follows these steps:
- It requires a bounding box to be drawn by the user to roughly segment out the foreground pixels
- It runs an initial min-cut optimization using the provided annotation
- The result of this optimization gives an initial segmentation 
- To further refine this segmentation, the user provides two kinds of strokes to aid the optimization
    - strokes on the background pixels
    - strokes on the foreground pixels
- The algorithm now utilizes this to refine the original segmentation

You are allowed to use standard GMM libraries for the implementation. For usage of other libraries, please contact the TAs.

You can view this [video](https://www.youtube.com/watch?v=aOqOwM-Qbtg) to get a better idea of the steps involved.

Image segmentation is one exciting application of MRFs. You can further read about other applications of MRFs for Computer Vision [here](https://cedar.buffalo.edu/~srihari/CSE574/Chap8/Ch8-PGM-Undirected/9.5-MRFinCV.pdf).

# Deliverable 1 (Algorithm)

```python
import numpy as np
import cv2
import networkx as nx
import igraph as ig
import shutil
import sys
import time
import matplotlib.pyplot as plt
from pathlib import Path
from sklearn.mixture import GaussianMixture
```

### Event Handler

```python
class EventHandler:
    """
    Class for handling user input during segmentation iterations
    """

    def __init__(self, flags, img, _mask, _umask, colors):

        self.FLAGS = flags
        self.ix = -1
        self.iy = -1
        self.img = img
        self.img2 = self.img.copy()
        self._mask = _mask
        self._umask = _umask
        self.COLORS = colors

    @property
    def image(self):
        return self.img

    @image.setter
    def image(self, img):
        self.img = img

    @property
    def mask(self):
        return self._mask

    @property
    def umask(self):
        return self._umask

    @mask.setter
    def mask(self, _mask):
        self._mask = _mask

    @umask.setter
    def umask(self, _umask):
        self._umask = _umask

    @property
    def flags(self):
        return self.FLAGS

    @flags.setter
    def flags(self, flags):
        self.FLAGS = flags

    def handler(self, event, x, y, flags, param):

        # Draw the rectangle first
        if event == cv2.EVENT_RBUTTONDOWN:
            self.FLAGS['DRAW_RECT'] = True
            self.ix, self.iy = x,y

        elif event == cv2.EVENT_MOUSEMOVE:
            if self.FLAGS['DRAW_RECT'] == True:
                self.img = self.img2.copy()
                cv2.rectangle(self.img, (self.ix, self.iy), (x, y), self.COLORS['BLUE'], 2)
                self.FLAGS['RECT'] = (min(self.ix, x), min(self.iy, y), abs(self.ix - x), abs(self.iy - y))
                self.FLAGS['rect_or_mask'] = 0

        elif event == cv2.EVENT_RBUTTONUP:
            self.FLAGS['DRAW_RECT'] = False
            self.FLAGS['rect_over'] = True
            cv2.rectangle(self.img, (self.ix, self.iy), (x, y), self.COLORS['BLUE'], 2)
            self.FLAGS['RECT'] = (min(self.ix, x), min(self.iy, y), abs(self.ix - x), abs(self.iy - y))
            self.FLAGS['rect_or_mask'] = 0

            x, y, w, h = self.FLAGS['RECT']
            self._mask[y:y+h,x:x+w] = 1
            tmp = np.zeros_like(self._mask)
            tmp[y:y+h,x:x+w] = 1
            self._umask = 1 - tmp

        # Draw strokes for refinement
        if event == cv2.EVENT_LBUTTONDOWN:
            if self.FLAGS['rect_over'] == False:
                print('Draw the rectangle first.')
            else:
                self.FLAGS['DRAW_STROKE'] = True
                cv2.circle(self.img, (x,y), 3, self.FLAGS['value']['color'], -1)
                cv2.circle(self._mask, (x,y), 3, self.FLAGS['value']['val'], -1)
                cv2.circle(self._umask, (x,y), 3, 1, -1)

        elif event == cv2.EVENT_MOUSEMOVE:
            if self.FLAGS['DRAW_STROKE'] == True:
                cv2.circle(self.img, (x, y), 3, self.FLAGS['value']['color'], -1)
                cv2.circle(self._mask, (x, y), 3, self.FLAGS['value']['val'], -1)
                cv2.circle(self._umask, (x,y), 3, 1, -1)

        elif event == cv2.EVENT_LBUTTONUP:
            if self.FLAGS['DRAW_STROKE'] == True:
                self.FLAGS['DRAW_STROKE'] = False
                cv2.circle(self.img, (x, y), 3, self.FLAGS['value']['color'], -1)
                cv2.circle(self._mask, (x, y), 3, self.FLAGS['value']['val'], -1)
                cv2.circle(self._umask, (x,y), 3, 1, -1)
```

### Metrics

```python
def calc_acc(predicted, ground_truth):
    return np.mean(predicted == ground_truth)

def calc_iou(predicted, ground_truth):
    inter = np.sum(np.logical_and(predicted, ground_truth))
    union = np.sum(np.logical_or(predicted, ground_truth))
    return inter / union

def calc_dice(predicted, ground_truth):
    inter = 2*np.sum(np.logical_and(predicted, ground_truth))
    union = np.sum(predicted > 0) + np.sum(ground_truth > 0)
    return inter / union

def calc_metrics(predicted, ground_truth):
    return calc_acc(predicted, ground_truth), \
            calc_iou(predicted, ground_truth), \
            calc_dice(predicted, ground_truth)
```

### Run Loop

```python
def run(filename: str, gt_filename: str, file: str, save: bool=False, components=5, neighbours=8, gamma=100):
    """
    Main loop that implements GrabCut.

    Input
    -----
    filename (str) : Path to image
    """
    # To keep track of iterations
    iter = 0
    if save:
        # Create ouput directory
        op_path = Path(f"../outputs/{file}")
        shutil.rmtree(op_path, ignore_errors=True)
        op_path.mkdir(parents=True, exist_ok=True)

    eng = []
    masks = []

    COLORS = {
    'BLACK' : [0,0,0],
    'RED'   : [0, 0, 255],
    'GREEN' : [0, 255, 0],
    'BLUE'  : [255, 0, 0],
    'WHITE' : [255,255,255]
    }

    DRAW_BG = {'color' : COLORS['BLACK'], 'val' : 0}
    DRAW_FG = {'color' : COLORS['WHITE'], 'val' : 1}

    FLAGS = {
        'RECT' : (0, 0, 1, 1),
        'DRAW_STROKE': False,         # flag for drawing strokes
        'DRAW_RECT' : False,          # flag for drawing rectangle
        'rect_over' : False,          # flag to check if rectangle is  drawn
        'rect_or_mask' : -1,          # flag for selecting rectangle or stroke mode
        'value' : DRAW_FG,            # drawing strokes initialized to mark foreground
    }

    img = cv2.imread(filename)
    img2 = img.copy()
    mask = np.zeros(img.shape[:2], dtype = np.uint8) # mask is a binary array with : 0 - background pixels
                                                     #                               1 - foreground pixels
    umask = np.zeros(img.shape[:2], dtype = np.uint8) # umask is a binary array with : 0 - background pixels
                                                      #                               1 - foreground pixels
    output = np.zeros(img.shape, np.uint8)           # output image to be shown

    # Input and segmentation windows
    # cv2.namedWindow('Input Image')
    # cv2.namedWindow('Segmented output')
    cv2.namedWindow('Input Image', cv2.WINDOW_GUI_NORMAL)
    cv2.namedWindow('Segmented output', cv2.WINDOW_GUI_NORMAL)

    EventObj = EventHandler(FLAGS, img, mask, umask, COLORS)
    cv2.setMouseCallback('Input Image', EventObj.handler)
    cv2.moveWindow('Input Image', img.shape[1] + 10, 90)

    # Initialize GMM for bg and fg
    bg_gmm = None
    fg_gmm = None

    while(1):
        img = EventObj.image
        mask = EventObj.mask
        umask = EventObj.umask
        FLAGS = EventObj.flags
        cv2.imshow('Segmented output', output)
        cv2.imshow('Input Image', img)

        k = cv2.waitKey(1)

        # key bindings
        if k == 27:
            # esc to exit
            break

        elif k == ord('0'):
            # Strokes for background
            FLAGS['value'] = DRAW_BG

        elif k == ord('1'):
            # FG drawing
            FLAGS['value'] = DRAW_FG

        elif k == ord('r'):
            # reset everything
            FLAGS['RECT'] = (0, 0, 1, 1)
            FLAGS['DRAW_STROKE'] = False
            FLAGS['DRAW_RECT'] = False
            FLAGS['rect_or_mask'] = -1
            FLAGS['rect_over'] = False
            FLAGS['value'] = DRAW_FG
            img = img2.copy()
            mask = np.zeros(img.shape[:2], dtype = np.uint8)
            umask = np.zeros(img.shape[:2], dtype = np.uint8)
            EventObj.image = img
            EventObj.mask = mask
            EventObj.umask = umask
            output = np.zeros(img.shape, np.uint8)

        elif k == 13:
            # Update GMM
            print(f"-- Iteration: {iter+1}")
            print("-- Fitting GMM")
            t0 = time.time()
            
            # Fit GMM
            bg_gmm, fg_gmm = fit_gmm(img2, mask, bg_gmm, fg_gmm, components)

            # Update graph
            print("-- Creating graph")
            graph = create_graph(img2, mask, umask, bg_gmm, fg_gmm, neighbours, gamma)
            
            # Find mincut
            print("-- Running graph cut")
            value, (fg, bg) = grab_cut(graph, img)
            
            # Update mask
            for idx in fg:
                y = idx // img.shape[1]
                x = idx % img.shape[1]
                try:
                    mask[y][x] = 1
                except IndexError:
                    pass
            for idx in bg:
                y = idx // img.shape[1]
                x = idx % img.shape[1]
                try:
                    mask[y][x] = 0
                except IndexError:
                    pass
                
            masks.append(mask.copy())
            eng.append(value)
            
            # Calculate metrics
            gt = cv2.imread(gt_filename, 0) // 255
            acc, iou, dice = calc_metrics(mask, gt)
            
            # Print summary
            print(f"-- Done. Time Taken: {time.time() - t0:.2f}s")
            print(f"-- Max Flow: {value:.2f}")
            print(f"-- Acc: {acc:.2f}, IoU: {iou:.2f}, Dice: {dice:.2f}")
            print(50*"-")

            # Save output
            if save:
                cv2.imwrite(str(op_path / f"{iter}.bmp"), mask*255)
            iter += 1

        EventObj.flags = FLAGS
        mask2 = np.where((mask == 1), 255, 0).astype('uint8')
        output = cv2.bitwise_and(img2, img2, mask = mask2)

    return masks, eng
```

### Create Graph

```python
def create_graph(img, mask, umask, bg_gmm, fg_gmm, neighbours=8, gamma=100):
    # Create graph
    height, width = img.shape[:2]
    graph = nx.lattice.grid_2d_graph(height, width)
    if neighbours == 8:
        graph.add_edges_from(
            [
                ((x, y), (x + 1, y + 1))
                for x in range(height - 1)
                for y in range(width - 1)
            ]
            +
            [
                ((x + 1, y), (x, y + 1))
                for x in range(height - 1)
                for y in range(width - 1)
            ])

    # Find beta
    wts = []
    edges = []
    for edge in graph.edges:
        src = edge[0]
        dst = edge[1]
        wt = np.sum((img[src] - img[dst])**2)
        wts.append(wt)
        edges.append((src, dst))
    beta = 1/(np.mean(wts)*2)

    # Find edge potentials
    for i in range(len(edges)):
        src, dst = edges[i]
        graph[src][dst]['weight'] = gamma * np.exp(-beta*wts[i])

    # Add unary potentials
    source = (height, width)
    sink = (height+1, width+1)

    # Unknown
    ys, xs = np.nonzero(umask == 0)
    bg_preds = -bg_gmm.score_samples(img[ys,xs])
    fg_preds = -fg_gmm.score_samples(img[ys,xs])
    for i in range(len(ys)):
        graph.add_edge(source, (ys[i], xs[i]), weight=bg_preds[i])
        graph.add_edge(sink, (ys[i], xs[i]), weight=fg_preds[i])

    # Foreground
    ys, xs = np.nonzero(np.logical_and(mask == 1, umask == 1))
    for i in range(len(ys)):
        graph.add_edge(source, (ys[i], xs[i]), weight=9*gamma)

    # Background
    ys, xs = np.nonzero(np.logical_and(mask == 0, umask == 1))
    for i in range(len(ys)):
        graph.add_edge(sink, (ys[i], xs[i]), weight=9*gamma)

    return graph
```

### Fit GMM

```python
def fit_gmm(img, mask, bg_gmm=None, fg_gmm=None, components=5):
    if not bg_gmm:
        bg_gmm = GaussianMixture(n_components=components)
    bg_gmm.fit(img[mask == 0])

    if not fg_gmm:
        fg_gmm = GaussianMixture(n_components=components)

    fg_gmm.fit(img[mask == 1])

    return bg_gmm, fg_gmm
```

### Graph Cut

```python
def grab_cut(graph, img):
    # Convert 2d indices to 1d
    height, width = img.shape[:2]
    x, y = np.meshgrid(np.arange(height), np.arange(width))
    grid = np.dstack((x, y)).reshape(-1, 2)
    mapping = {tuple(coord.tolist()): coord[0]*width + coord[1] for coord in grid}
    mapping[(height, width)] = height*width 
    mapping[(height+1, width+1)] = height*width+1
    graph = nx.relabel_nodes(graph, mapping)

    # Convert networkx graph to igraph as its faster
    edges = []
    wts = []
    for edge in graph.edges:
        edges.append(edge)
        wts.append(graph.get_edge_data(edge[0], edge[1])['weight'])

    g = ig.Graph(directed=False)
    g.add_vertices(height*width + 2)
    g.add_edges(edges, attributes={'weight': wts})

    result = g.st_mincut(height*width, height*width+1, capacity='weight')
    return result.value, result.partition
```

```python
def present(file):
    filename = f"../images/{file}.jpg"
    gt = f"../ground_truth/{file}.bmp"
    img = cv2.cvtColor(cv2.imread(filename), cv2.COLOR_BGR2RGB)
    gt_mask = cv2.imread(gt, cv2.IMREAD_GRAYSCALE)

    masks, eng = run(filename, gt, file)
    cv2.destroyAllWindows()

    fig, axs = plt.subplots(len(masks)+1, 2, figsize=(9, 4*len(masks)+1))
    axs[0,0].imshow(gt_mask, cmap='gray')
    axs[0,0].axis('off')
    axs[0,0].title.set_text('Mask (Ground Truth)')
    output = img.copy()
    output[gt_mask == 0] = 0
    axs[0,1].imshow(output)
    axs[0,1].axis('off')
    axs[0,1].title.set_text(f'Image (Ground Truth)')
    for i in range(len(masks)):
        mask = np.where((masks[i] == 1), 255, 0).astype('uint8')
        output = cv2.bitwise_and(img, img, mask=mask)
        axs[i+1,0].imshow(masks[i]*255, cmap='gray')
        axs[i+1,0].axis('off')
        axs[i+1,0].title.set_text(f'Mask (Iteration {i+1})')
        axs[i+1,1].imshow(output)
        axs[i+1,1].axis('off')
        axs[i+1,1].title.set_text(f'Image (Iteration {i+1})')
```

```python
gts = sorted(Path("../ground_truth").glob("*.bmp"))
results = []
names = []
print("Image,Accuracy,IoU,Dice\n")
for gt in gts:
    dirname = Path("../outputs") / gt.with_suffix('').name
    pred = cv2.imread(str(sorted(dirname.glob("*.bmp"))[-1]), 0) // 255
    gt_img = cv2.imread(str(gt), 0)
    acc, iou, dice = calc_metrics(pred, gt_img)
    names.append(dirname.name)
    results.append([acc, iou, dice])
    print(f"{dirname.name},{acc:.2f},{iou:.2f},{dice:.2f}")

results = np.asarray(results)
print("\nMean values:", np.mean(results, axis=0))
```
